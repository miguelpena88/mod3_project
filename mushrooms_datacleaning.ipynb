### Dummy Dataframe

```python
#creating variable for list of columns
cols = list(df.columns)
#creating dummies dataframe from our main df
df_dummies = pd.get_dummies(df, columns=cols)
```

```python
#checking list of dummy columns in df
list(df_dummies.columns)
```

```python
#creating a variable for list of dummy columns I will be dropping
dummy_drops= ['edible_poisonous_e','cap_shape_b','cap_surface_f','cap_color_b','bruises_f',
              'odor_a','gill_attachment_a','gill_spacing_c','gill_size_b','gill_color_b',
              'stalk_shape_e','stalk_root_b','stalk_surface_above_ring_f','stalk_surface_below_ring_f',
              'stalk_color_above_ring_b','stalk_color_below_ring_b','veil_color_n',
              'ring_number_n','ring_type_e','spore_print_color_b','population_a','habitat_d']
```

```python
#dropping a dummy column per dummy set from df_dummies
df_dummies.drop(dummy_drops, axis=1, inplace=True)
```
